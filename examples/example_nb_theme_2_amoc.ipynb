#### **1.1 Creating NEMODataTrees**
---

* Let's begin by importing the **NEMO Cookbook** Python library and the `NEMODataTree` class:

```python
import xarray as xr

from nemo_cookbook import NEMODataTree
from OceanDataStore import OceanDataCatalog
```

* Creating a `NEMODataTree` from the **eORCA1 ERA5v1** monthly-mean meridional velocity outputs using the **OceanDataCatalog**:

```python
catalog = OceanDataCatalog(catalog_name="noc-model-stac")
```

```python
# Open eORCA1 ERA5 v1 domain_cfg ancillary dataset:
ds_domain_cfg = catalog.open_dataset(id="noc-npd-era5/npd-eorca1-era5v1/gn/domain/domain_cfg")
# Open eORCA1 ERA5 v1 gridV dataset:
ds_gridV = catalog.open_dataset(id="noc-npd-era5/npd-eorca1-era5v1/gn/V1m")

ds_gridV
```

```python
# Define datasets dictionary:
datasets = {"parent": {"domain": ds_domain_cfg, "gridV": ds_gridV}}

# Create NEMODataTree from eORCA1 ERA5 v1 datasets:
nemo_eorca1 = NEMODataTree.from_datasets(datasets=datasets, iperio=True, nftype="F", read_mask=True)

nemo_eorca1
```

    * Overturning in the Subpolar North Atlantic Program (**OSNAP**) array
    * Rapid Climate Change-Meridional Overturning Circulation and Heatflux Array (**RAPID-MOCHA**) at 26.5Â°N
    * Meridional Overturning Variability Experiment (**MOVE**) array at 16Â°N
    * South Atlantic Meridional overturning circulation Basin-wide Array (**SAMBA**) array at 34.5Â°S

* The latter three diagnostic datasets are calculated using the Meridional ovErTurning ciRculation diagnostIC ([**METRIC**](https://github.com/AMOCcommunity/metric)) package.

* We can access these diagnostics by searching the **OceanDataCatalog** for any *Items* with identifiers containing the key word `"tn"`, which corresponds to transects which are defined on the native NEMO model grid:

```python
catalog.search(collection='noc-npd-era5', item_name='tn')
```

#### **1.3 Accessing AMOC Observations**
---

* We can access [**RAPID 26.5N**](https://rapid.ac.uk/data) AMOC observations stored in the JASMIN object store using `xarray.open_zarr()`:

```python
# Define URL to RAPID Zarr dataset:
url = "https://noc-msm-o.s3-ext.jc.rl.ac.uk/ocean-obs/RAPID/moc_transports_v2024.1/"

# Open RAPID dataset:
ds_rapid = xr.open_zarr(url, zarr_format=3)

ds_rapid
```

* We can access [**Overturning in the Subpolar North Atlantic (OSNAP)**](https://www.o-snap.org/data-access/) AMOC observations stored in the JASMIN object store using `xarray.open_zarr()`:

```python
# Define URL to OSNAP Zarr dataset:
url = "https://noc-msm-o.s3-ext.jc.rl.ac.uk/ocean-obs/OSNAP/OSNAP_MOC_MHT_MFT_TimeSeries_201408_202207_2025/"

# Open OSNAP dataset:
ds_osnap = xr.open_zarr(url, zarr_format=3)

ds_osnap
```
