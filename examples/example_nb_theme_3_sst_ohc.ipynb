#### **1.1 Creating NEMODataTrees**
---

* Let's begin by importing the **NEMO Cookbook** Python library and the `NEMODataTree` class:

```python
import xarray as xr

from nemo_cookbook import NEMODataTree
from OceanDataStore import OceanDataCatalog
```

* Creating a `NEMODataTree` from the **eORCA1 ERA5v1** monthly-mean sea surface temperature outputs using the **OceanDataCatalog**:

```python
# Initialising the OceanDataCatalog using the `noc-model-stac` stored in the JASMIN cloud object store:
catalog = OceanDataCatalog(catalog_name="noc-model-stac")

# Alternative: Initialising the OceanDataCatalog using the `noc-model-stac` stored locally:
# catalog = OceanDataCatalog(catalog_url="/groups/npd-workshop/noc-model-stac/catalog.json")
```

```python
# Open eORCA1 ERA5 v1 domain_cfg ancillary dataset:
ds_domain_cfg = catalog.open_dataset(id="noc-npd-era5/npd-eorca1-era5v1/gn/domain/domain_cfg")
# Open eORCA1 ERA5 v1 gridT dataset:
ds_gridT = catalog.open_dataset(id="noc-npd-era5/npd-eorca1-era5v1/gn/T1m")

ds_gridT
```

```python
# Define datasets dictionary:
datasets = {"parent": {"domain": ds_domain_cfg, "gridT": ds_gridT}}

# Create NEMODataTree from eORCA1 ERA5 v1 datasets:
nemo_eorca1 = NEMODataTree.from_datasets(datasets=datasets, iperio=True, nftype="F", read_mask=True)

nemo_eorca1
```

#### **1.2 Accessing Sea Surface Temperature Observations**
---

* We can access [**OISSTv2**](https://psl.noaa.gov/data/gridded/data.noaa.oisst.v2.highres.html) sea surface temperature observations stored in the JASMIN object store using `xarray.open_zarr()`:

```python
# Define URL to OISST version 2 Zarr dataset:
url = "https://noc-msm-o.s3-ext.jc.rl.ac.uk/ocean-obs/OISSTv2/OISSTv2_sst_global_monthly/"

# Open OISST v2 dataset:
ds_oisst = xr.open_zarr(url, zarr_format=3)

ds_oisst
```

* We can access [**HadISST**](https://www.metoffice.gov.uk/hadobs/hadisst/) sea surface temperature observations stored in the JASMIN object store using `xarray.open_zarr()`:

```python
# Define URL to HadISST Zarr dataset:
url = "https://noc-msm-o.s3-ext.jc.rl.ac.uk/ocean-obs/HadISST/HadISST_global_monthly/"

# Open HadISST v2 dataset:
ds_hadisst = xr.open_zarr(url, zarr_format=3)

ds_hadisst
```
