#### **1.1 Creating NEMODataTrees**
---

* Let's begin by importing the **NEMO Cookbook** Python library and the `NEMODataTree` class:

```python
import xarray as xr

from nemo_cookbook import NEMODataTree
from OceanDataStore import OceanDataCatalog
```

* Creating a `NEMODataTree` from the **eORCA1 ERA5v1** monthly-mean sea ice outputs using the **OceanDataCatalog**:

```python
catalog = OceanDataCatalog(catalog_name="noc-model-stac")
```

```python
# Open eORCA1 ERA5 v1 domain_cfg ancillary dataset:
ds_domain_cfg = catalog.open_dataset(id="noc-npd-era5/npd-eorca1-era5v1/gn/domain/domain_cfg")
# Open eORCA1 ERA5 v1 icemod dataset:
ds_icemod = catalog.open_dataset(id="noc-npd-era5/npd-eorca1-era5v1/gn/I1m")

ds_icemod
```

```python
# Define datasets dictionary:
datasets = {"parent": {"domain": ds_domain_cfg, "icemod": ds_icemod}}

# Create NEMODataTree from eORCA1 ERA5 v1 datasets:
nemo_eorca1 = NEMODataTree.from_datasets(datasets=datasets, iperio=True, nftype="F", read_mask=True)

nemo_eorca1
```

#### **1.2 Accessing Sea Ice Observations**
---

* We can access [**NSIDC**](https://nsidc.org/data/g02135/versions/4) sea ice observations stored in the JASMIN object store using `xarray.open_zarr()`:

```python
# Define URL to NSIDC Sea Ice Index version 4 Zarr dataset:
url = "https://noc-msm-o.s3-ext.jc.rl.ac.uk/ocean-obs/NSIDC/NSIDC_Sea_Ice_Index_v4_Arctic/"
# url = "https://noc-msm-o.s3-ext.jc.rl.ac.uk/ocean-obs/NSIDC/NSIDC_Sea_Ice_Index_v4_Antarctic/"

# Open NSIDC Sea Ice dataset:
ds_nsidc = xr.open_zarr(url, zarr_format=3)

ds_nsidc
```

* We can access [**HadISST**](https://www.metoffice.gov.uk/hadobs/hadisst/) sea ice observations stored in the JASMIN object store using `xarray.open_zarr()`:

```python
# Define URL to HadISST monthly sea ice Zarr dataset:
url = "https://noc-msm-o.s3-ext.jc.rl.ac.uk/ocean-obs/HadISST/HadISST_global_monthly/"

# Open HadISST monthly sea ice dataset:
ds_hadisst = xr.open_zarr(url, zarr_format=3)

ds_hadisst
```
